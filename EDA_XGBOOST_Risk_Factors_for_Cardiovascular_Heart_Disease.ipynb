# Data Wrangling

As there is no such thing as negative pressure, let's assume this as mistake and all the negative values convert to positive ones.

```python
df['ap_hi_pos'] = df['ap_hi'].abs()
df['ap_lo_pos'] = df['ap_lo'].abs()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,5))
sns.boxplot(df['ap_hi_pos'], ax=ax1)
sns.boxplot(df['ap_lo_pos'], ax=ax2)
```

```python
df['ap_hi_pos'].describe(), df['ap_lo_pos'].describe()
```

Then let's assume the values in format XXXXX or XXXX are the values XXX.XX or XX.XX respectively, so I convert all the 4 a 5 digits data into numbers with 2 decimal places. 

```python
df['ap_hi_pos'].loc[df['ap_hi_pos']<=70].count()
```

```python
df['ap_hi_pos'].loc[df['ap_hi_pos']>=1000].count()
```

```python
df['ap_hi_pos'].astype(str).str.len().value_counts()
```

```python
def decimals(item):
    if len(str(item)) == 4:
        item = item/10
    if len(str(item)) == 5:
        item = item/100
    return item
```

```python
df['ap_hi_clean'] = df['ap_hi_pos'].apply(np.vectorize(decimals))
```

```python
df['ap_lo_clean'] = df['ap_lo_pos'].apply(np.vectorize(decimals))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,5))
sns.boxplot(df['ap_hi_clean'], ax=ax1)
sns.boxplot(df['ap_lo_clean'], ax=ax2)
```

Obviously the data are still not perfect, but there is no source to check this out. I will keep them for purpose of this notebook as they are.

Let's scale the data to avoid working with high numbers. 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
data_to_scale = ['age','height', 'weight', 'ap_hi_clean', 'ap_lo_clean']
```

```python
df[data_to_scale] = scaler.fit_transform(df[data_to_scale])
```

```python
df[data_to_scale]
```

```python
data_to_ML = ['age','height', 'weight','ap_hi_clean', 'ap_lo_clean', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio']
```

```python
train, test = train_test_split(df[data_to_ML], test_size=0.2,random_state=0)
```

```python
y_train = train['cardio']
X_train = train.drop("cardio", axis=1)

y_test = test['cardio']
X_test = test.drop('cardio', axis=1)
```

```python
X_train.shape, y_train.shape
```

# XGBoost (extra gradient boosting ensamble model)

```python
xgbModel = xgb.XGBClassifier(
    n_estimators=1000,
    reg_lambda=1,
    gamma=0,
    max_depth=4
)
 
xgbModel.fit(X_train, y_train,
             verbose = False)
```

```python
important_features = pd.DataFrame(xgbModel.feature_importances_.reshape(1, -1), columns=X_train.columns)

important_features
```

It seems the most significant feature is systolic blood pressure (ap_hi_clean feature). 

```python
y_pred = xgbModel.predict(X_test)
 
xgbModel.score(X_test, y_test)
```

The accuracy level is only 72% - too low for ML model. We have to adjust the model or wrangle and clean the data much more. There are still many outliers without explanation that can influence the final model.
